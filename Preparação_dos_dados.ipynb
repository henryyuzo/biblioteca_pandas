```python
df_i = pd.read_csv('/content/drive/MyDrive/Cursos/Escola DNC/Formação de dados/Mês 3/Semana 9/invested.csv')
```

```python
df_i = pd.get_dummies(df_i, columns = ['backedLocation'])
```

```python
df_i.head()
```

```python
df_i = df_i.rename(columns={'backedLocation_BR' : 'BR',	'backedLocation_GBK' : 'GBK',	'backedLocation_US' : 'US'})
```

```python
df_i.head()
```

```python
df_i.sort_values(by = ['ID'])
```

```python
df_i = df_i.groupby(by = ['ID']).agg({'age' : 'mean', 'BR' : 'sum','GBK' : 'sum', 'US' : 'sum'})
```

```python
df_i.head()
```
